```python
# Check for missing values in the dataset
null_counts = df.isnull().sum()
# Calculate percentage of missing values
null_percentage = (
    null_counts / len(df) * 100
).round(2)
# Create a summary table
null_summary = pd.DataFrame({
    'Missing Values': null_counts,
    'Percentage (%)': null_percentage
})
# Show only columns with missing values
null_summary = null_summary[
    null_summary['Missing Values'] > 0
]

print("Missing Value Analysis:\n")

display(null_summary)
```

```python
print(df.columns.tolist())
```

```python
# Handle missing values

# Fill missing seller names
df['seller_name'] = df['seller_name'].fillna('Unknown')

# Fill missing ratings with median
df['rating'] = df['rating'].fillna(
    df['rating'].median()
)

# Fill missing ratings count with median
df['ratings_count'] = df['ratings_count'].fillna(
    df['ratings_count'].median()
)

# Remove rows where title is missing
df = df.dropna(subset=['title'])

print("Missing values handled successfully!")

print(df.isnull().sum())
```

```python
# Remove duplicate rows from the dataset

before = len(df)
df = df.drop_duplicates()
after = len(df)
print("Duplicate rows removed successfully!")

print("Rows before cleaning :", before)
print("Rows after cleaning  :", after)
print("Duplicates removed   :", before - after)
```

```python
# Clean price columns

def clean_price(series):

    return (
        series.astype(str)
        .str.replace('₹', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
        .astype(float)
    )
# Convert price columns to numeric format
df['initial_price'] = clean_price(df['initial_price'])
df['final_price'] = clean_price(df['final_price'])
print("Price columns cleaned successfully!")
print("\nFirst 3 rows:")
print(df[['initial_price', 'final_price']].head(3))
```

```python
# Select important columns for analysis

useful_cols = [
    'product_id',
    'title',
    'initial_price',
    'final_price',
    'rating',
    'ratings_count'
]
df = df[useful_cols].copy()
# Filter products with rating 4 or above
df_filtered = df[df['rating'] >= 4]
print("Filtered Products :", len(df_filtered))
print("Remaining Products:", len(df) - len(df_filtered))
# Display first few filtered records
df_filtered.head(10)
```

```python
# Remove duplicate records
before = len(df)
# Remove duplicate rows
df = df.drop_duplicates()
after = len(df)
print("Duplicate rows removed successfully!")
print("Rows before :", before)
print("Rows after  :", after)
print("Duplicates removed :", before - after)
```

```python
# Create derived columns

# Generate random quantity values between 1 and 5
np.random.seed(42)

df_filtered['quantity'] = np.random.randint(
    1,
    6,
    size=len(df_filtered)
)
# Calculate total amount
df_filtered['total_amount'] = (
    df_filtered['final_price'] *
    df_filtered['quantity']
)
print("Derived columns added successfully!")
# Display first 5 rows
# Display first 5 rows with 10 columns

df_filtered[
    [
        'product_id',
        'title',
        'initial_price',
        'final_price',
        'rating',
        'ratings_count',
        'quantity',
        'total_amount'
    ]
].head()
```

```python
# Generate summary statistics

summary = {
    "Total Filtered Products"     : len(df_filtered),
    "Average Final Price (₹)"     : round(df_filtered['final_price'].mean(), 2),
    "Average Rating"              : round(df_filtered['rating'].mean(), 2),
    "Total Revenue (₹)"           : round(df_filtered['total_amount'].sum(), 2),
    "Average Revenue per Product" : round(df_filtered['total_amount'].mean(), 2),
}

print("Summary Statistics\n" + "-" * 40)

for key, value in summary.items():
    print(f"{key:30}: {value}")
```

```python
# Export cleaned dataset

OUTPUT_PATH = r"C:\Users\vidiy\OneDrive\Desktop\1st\cleaned_dataset.csv"

try:

    df_filtered.to_csv(
        OUTPUT_PATH,
        index=False
    )

    print("Cleaned dataset exported successfully!")

    print(f"\nSaved file location : {OUTPUT_PATH}")

    print(
        f"Final Dataset Shape : "
        f"{df_filtered.shape[0]} rows × "
        f"{df_filtered.shape[1]} columns"
    )

except Exception as e:

    print(f"Error saving file: {e}")
```
